```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
# Line plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label = 'Sin(x)', color = 'blue', linestyle = '-', linewidth = 3)
plt.title("Line plot of sine wave")
plt.xlabel("X values")
plt.ylabel("Sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
