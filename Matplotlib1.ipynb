```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
# Line plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label = 'Sin(x)', color = 'aqua', linestyle = '-.', linewidth = 3)
plt.title("Line plot of sine wave")
plt.xlabel("X values")
plt.ylabel("Sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plot
np.random.seed(0)
x = np.random.rand(40)
y = np.random.rand(40)

plt.figure(figsize=(6,4))
plt.scatter(x, y, c = 'red', marker = 'o', alpha = 0.9)
plt.title("Scatter plot")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Bar plot
categories = ['A','B','C','D']
values = [20,50,10,70]
plt.figure(figsize=(6,4))
plt.bar(categories, values, color='green'
```
